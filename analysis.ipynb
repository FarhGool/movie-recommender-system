```python
import numpy as np
import pandas as pd
import ast
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle

```

```python
# Merge both datasets on title
movies = movie.merge(credits, on='title')
```

```python
movies.head()
```

```python
# Keep important columns
movies = movies[['id', 'title', 'overview', 'keywords', 'genres', 'cast', 'crew']]

# Drop rows with null values
movies.dropna(inplace=True)
```

```python
movies.head()
```

```python
# Process 'overview' column
movies['overview'] = movies['overview'].apply(lambda x: x.split())

```

```python
# Process 'keywords' column
def convert_keywords(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l

movies['keywords'] = movies['keywords'].apply(convert_keywords)
```

```python
# Process 'genres' column
def convert_genres(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l

movies['genres'] = movies['genres'].apply(convert_genres)
```

```python
# Process 'cast' column (top 3 only)
def extract_cast(obj):
    l = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            l.append(i['name'])
            count += 1
        else:
            break
    return l

movies['cast'] = movies['cast'].apply(extract_cast)
```

```python
# Process 'crew' column (get director only)
def extract_director(obj):
    l = []
    try:
        for i in ast.literal_eval(obj):
            if i['job'] == 'Director':
                l.append(i['name'])
                break
    except Exception as e:
        print("Error:", obj[:100])  # print first 100 chars
        return []
    return l

movies['crew'] = movies['crew'].apply(extract_director)
```

```python

# Tags = Combined features of a movie
# Create 'tags' column by combining overview + keywords + genres + cast + crew
movies['tags'] = movies['overview'] + movies['cast'] + movies['crew'] + movies['keywords'] + movies['genres']
# Remove spaces from tags
movies['tags'] = movies['tags'].apply(lambda x: [i.replace(" ", "") for i in x])

# Stemming : Reduces words to their base form. Example: Loved, Loving, Loves -> Love
ps = PorterStemmer()

def stemming(text):
    l = []
    for i in text:
        l.append(ps.stem(i))
    return " ".join(l)

movies['tags'] = movies['tags'].apply(stemming)


movies['tags'].head()
```

```python
# Final dataset with relevant columns
movies = movies[['id', 'title', 'tags']]
movies.head()
```

```python
# Vectorization: Convert text data into numerical data using CountVectorizer
vectorizer = CountVectorizer(max_features=500, stop_words='english')
vectors = vectorizer.fit_transform(movies['tags']).toarray()

# Cosine similarity: Measure similarity between two vectors. Range: 0 to 1. Closer to 1 means more similar.
similarity = cosine_similarity(vectors)
```

```python
# Recommendation function:
'''
def Recommendation_system(movie_title):
    movie_index = movies[movies['title'] == movie_title].index[0] # Get index of the movie
    distances = sorted(list(enumerate(similarity[movie_index])), reverse=True, key=lambda x: x[1]) # Sort movies based on similarity score

    for i in distances[1:20]: # Loop through the top 20 similar movies (excluding the given movie itself)
        print(movies.iloc[i[0]].title) # Print the titles of the top 20 similar movies (excluding the given movie itself)
'''

def recommend(movie_title):
    movie_index = movies[movies['title'] == movie_title].index[0] # Get index of the movie
    distances = similarity[movie_index] # Get similarity scores for all movies with the given movie

    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6] # Get top 5 similar movies (excluding the given movie itself)

    recommended = [] # Initialize an empty list to store recommended movie titles
    for i in movies_list: # Loop through the list of similar movies
        recommended.append(movies.iloc[i[0]].title) # Append the title of the recommended movie to the list

    return recommended

# Example usage
# Recommendation_system('Avatar')
```

```python
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import KNNBasic
from surprise import accuracy
```

Due to lack of real user interaction data in the TMDB dataset, synthetic users were generated and movie vote averages were used as proxy ratings to enable collaborative filtering.

```python
# Create Ratings Dataset

movie1 = pd.read_csv('tmdb_5000_movies.csv')

# Use original movie dataframe (before dropping columns if possible)
ratings_df = movie1[['id', 'vote_average']].copy()

# Rename columns
ratings_df.columns = ['movieId', 'rating']

# Create synthetic users
num_users = 1000
ratings_df['userId'] = np.random.randint(1, num_users+1, size=len(ratings_df))

ratings_df = ratings_df[['userId', 'movieId', 'rating']]

ratings_df.head()
```

```python
# Load Data into Surprise

reader = Reader(rating_scale=(0, 10))

data = Dataset.load_from_df(
    ratings_df[['userId', 'movieId', 'rating']],
    reader
)

trainset, testset = train_test_split(data, test_size=0.2)
```

```python
# Train User-Based Collaborative Model

model = KNNBasic(sim_options={
    'name': 'cosine',
    'user_based': True
})

model.fit(trainset)
```

```python
pickle.dump(model, open('collab_model.pkl', 'wb'))
pickle.dump(ratings_df, open('ratings_df.pkl', 'wb'))
```

```python
# Evaluate Model

predictions = model.test(testset)
accuracy.rmse(predictions)
```

The model achieved an RMSE of 1.16, indicating moderate prediction error, which is expected due to synthetic user generation and absence of real interaction data

```python
# Recommendation Function (Collaborative)

def recommend_collaborative(user_id, top_n=5):

    movie_ids = ratings_df['movieId'].unique()
    predictions = []

    for movie_id in movie_ids:
        pred = model.predict(user_id, movie_id)
        predictions.append((movie_id, pred.est))

    # Sort by predicted rating
    predictions.sort(key=lambda x: x[1], reverse=True)

    top_movies = predictions[:top_n]

    # Convert to titles
    recommended = []
    for movie_id, score in top_movies:
        title = movies[movies['id'] == movie_id]['title'].values
        if len(title) > 0:
            recommended.append((title[0], score))

    return recommended
```

```python
# Test Collaborative Filtering

recommend_collaborative(user_id=1)
```

# Hybrid Methods

```python
def recommend_with_scores(movie_title, top_n=10):
    movie_index = movies[movies['title'] == movie_title].index[0]
    distances = similarity[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:top_n+1]

    results = []

    for i in movies_list:
        results.append({
            "movie_id": movies.iloc[i[0]].id,
            "title": movies.iloc[i[0]].title,
            "content_score": float(i[1])
        })

    return results
```

```python
# Hybrid Function (MAIN PART). we combine both models

def hybrid_recommend(user_id, movie_title, alpha=0.5, top_n=5):

    # Step 1: Get content-based recommendations
    content_results = recommend_with_scores(movie_title, top_n=20)

    hybrid_results = []

    for item in content_results:
        movie_id = item['movie_id']
        content_score = item['content_score']

        # Step 2: Collaborative score
        collab_score = model.predict(user_id, movie_id).est

        # Step 3: Normalize collaborative score (important!)
        collab_score_norm = collab_score / 10  # since ratings are 0–10

        # Step 4: Combine scores
        final_score = (alpha * content_score) + ((1 - alpha) * collab_score_norm)

        hybrid_results.append({
            "title": item['title'],
            "final_score": final_score
        })

    # Step 5: Sort results
    hybrid_results = sorted(
        hybrid_results,
        key=lambda x: x['final_score'],
        reverse=True
    )

    return hybrid_results[:top_n]
```

```python
# Test Hybrid System

hybrid_recommend(user_id=1, movie_title="Avatar")
```

Your hybrid system is working correctly because:
- It is producing ranked recommendations
- Scores are not identical → good sign (variation exists)
- Movies are relevant to Avatar (sci-fi / action genre overlap)
- Output is stable and sorted properly

The hybrid recommender successfully combines content-based similarity with user-based collaborative filtering using a weighted average approach. Results demonstrate strong thematic relevance, with slight ranking adjustments introduced by collaborative predictions. Due to synthetic user generation, collaborative influence is limited, resulting in a system that is primarily content-driven with collaborative refinement.

A hybrid recommendation system was implemented by combining content-based similarity scores with collaborative filtering predictions. A weighted average approach was used where both models contributed equally (α = 0.5). This ensures recommendations are influenced by both item similarity and user preference patterns, improving robustness and reducing limitations of individual approaches.
